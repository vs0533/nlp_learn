````python
from transformers import AutoTokenizer  #导入自动分词器

checkpoint = "./bert-base-chinese"
tokenizer=AutoTokenizer.from_pretrained('') #手动指定模型缓存的目录
````

````python
tokenizer
````

````python
sents = [
    "滚滚长江东逝水",
    "浪花淘尽英雄泪",
    "hello,world",
]
````

````python
encode = tokenizer(sents)
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

````python
encode = tokenizer(sents,padding=True,max_length=10)
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

````python
encode = tokenizer(sents,padding="max_length",max_length=10)
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

````python
encode = tokenizer(sents,padding="max_length",max_length=10,add_special_tokens=True)
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

````python
encode = tokenizer(sents,padding="max_length",max_length=10,add_special_tokens=True,return_tensors="pt")
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

````python
#sents = [("滚滚长江东逝水","浪花淘尽英雄泪"),("asdf","afff")]
sents = [("滚滚长江东逝水","浪花淘尽英雄泪"),("阿斯蒂芬","阿斯蒂芬付付付")]
encode = tokenizer(sents,padding="max_length",max_length=30,add_special_tokens=True,return_tensors="pt",truncation=True)
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

````python
sents = ["滚滚长江东逝水","浪花淘尽英雄泪","阿斯蒂芬","阿斯蒂芬付付付"]
encode = tokenizer.batch_encode_plus(
    batch_text_or_text_pairs=[
        (sents[0],sents[1]),
        (sents[2],sents[3]),
    ],
    padding="max_length",
    max_length=10,
    add_special_tokens=True,
    return_tensors="pt",
    truncation=True
)
print("encode:",encode)
print("input_ids:",encode["input_ids"])
for i in range(len(encode["input_ids"])):
    print(tokenizer.decode(encode["input_ids"][i]))
````

#### 加词 && 加符号

````python
tokenizer.add_tokens(new_tokens=["月光","希望"])
tokenizer.add_special_tokens({"eos_token":"[EOS]"})   #注意eos_token属性是bert模型规定的 不可自己定义 其值可以随便定义
tokenizer.add_special_tokens({"dd_token":"[BB]"})
````

````python
dict_info = tokenizer.get_vocab()
type(dict_info),len(dict_info),"月光" in dict_info
````

````python
text = "月光真美[EOS]"
text_encode = tokenizer(text)
text_decode = tokenizer.decode(text_encode["input_ids"])
text_decode      #可以发现”月光与<EOS>被当做了一个词进行处理“
````
